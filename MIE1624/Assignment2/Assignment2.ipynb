```python
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.preprocessing import OrdinalEncoder
from sklearn.feature_selection import chi2

%autosave 5
```

```python
response = pd.read_csv("clean_kaggle_data_2022.csv",low_memory=False)
response
```

```python
response = response.fillna("unknow")  ## Filling the empty with unknow

response = response.iloc[:,1:]  #drop the first column since no help for further analysis
response2 = response.iloc[1:,:]
response2
```

```python

#response2.drop("Q4",axis = 1)


def encode(response2,col):
    encode = {}
    i = 0

    for en in np.unique(response2[col]):
        if en != 'unknow':
            encode[en] = i
            i+=1
        else:
            encode[en] = 'unknow'
            i+=1
    return encode




```

```python
response2["Q26"].value_counts()
```

```python
Q11_encode = {  'I have never written code':0,
    '< 1 years': 1,
    '1-3 years': 2,
    '3-5 years': 3, 
    '5-10 years':4,
    '10-20 years': 5,
    '20+ years':6,
   }


Q43_encode = {  'Never': 0,
    'Once': 1,
    '2-5 times': 2, 
    '6-25 times':3,
    'More than 25 times': 4,
   }

platform_encode = encode(response2,col='Q32')

country_encode = encode(response2,col='Q4')

method_encode = encode(response2,col='Q27')


individual_encode = {  '0': 0,
    '1-2': 1,
    '3-4': 2, 
    '5-9':3,
    '10-14': 4,
    '15-19':5,
    '20+': 6
   }



num_empoly_encode = {   '0-49 employees': 0,
    '50-249 employees': 1,
    '250-999 employees': 2, 
    '1000-9,999 employees':3,
    '10,000 or more employees': 4,

   }

employe_encode = encode(response2,col='Q24')


role_encode = encode(response2,col="Q23")



hub_encode = encode(response2,col="Q22")

age_encode = {   '25-29': 0,
    '30-34': 1,
    '35-39': 1, 
    '40-44': 1,
    '22-24': 0,
    '45-49': 2, 
    '50-54': 2,
    '55-59': 2,
    '60-69': 3, 
    '18-21': 0,
    '70+': 3,
   }

gender_encode = {   'Man': 0,
    'Woman': 1,
    'Nonbinary': 2,
    'Prefer not to say': 3,
    'Prefer to self-describe': 4,
   }

Degree_encode = { 'Bachelor’s degree': 0,
    'Master’s degree': 1,
    'Doctoral degree': 2,
    'I prefer not to answer': 3,
    'Some college/university study without earning a bachelor’s degree': 4,
    'Professional doctorate': 5,
    'No formal education past high school ': 6,
}

Publish_encode = { 'Yes': 1,
    'No': 2,
    'unknow':'unknow'
}


experience_encode = {   'Under 1 year': 1,
    '1-2 years': 2,
    'I do not use machine learning methods': 0, 
    '2-3 years': 3,
    '5-10 years': 6,
    '4-5 years': 5, 
    '3-4 years': 4,
    '10-20 years': 7,
}


```

```python
response2["Q2"] = response2["Q2"].map(age_encode)
response2["Q3"] = response2["Q3"].map(gender_encode)
response2["Q4"] = response2["Q4"].map(country_encode)
response2["Q8"] = response2["Q8"].map(Degree_encode)
response2["Q9"] = response2["Q9"].map(Publish_encode)
response2["Q11"] = response2["Q11"].map(Q11_encode)
response2["Q16"] = response2["Q16"].map(experience_encode)
response2["Q22"] = response2["Q22"].map(hub_encode)
response2["Q23"] = response2["Q23"].map(role_encode)
response2["Q24"] = response2["Q24"].map(employe_encode)
response2["Q25"] = response2["Q25"].map(num_empoly_encode)
response2["Q26"] = response2["Q26"].map(individual_encode)
response2["Q27"] = response2["Q27"].map(method_encode)
response2["Q32"] = response2["Q32"].map(platform_encode)
response2["Q43"] = response2["Q43"].map(Q43_encode)
```
