### Combining the salary buckets and label encoding them

```python
salary_buckets = {   '$0-999': '0-9,999',
    '1,000-1,999': '0-9,999',
    '2,000-2,999': '0-9,999', 
    '3,000-3,999': '0-9,999',
    '4,000-4,999': '0-9,999',
    '5,000-7,499': '0-9,999', 
    '7,500-9,999': '0-9,999',
    '10,000-14,999': '10,000-19,999',
    '15,000-19,999': '10,000-19,999', 
    '20,000-24,999': '20,000-29,999',
    '25,000-29,999': '20,000-29,999',
    '30,000-39,999': '30,000-39,999', 
    '40,000-49,999': '40,000-49,999', 
    '50,000-59,999': '50,000-59,999',
    '60,000-69,999': '60,000-69,999',
    '70,000-79,999': '70,000-79,999', 
    '80,000-89,999': '80,000-89,999', 
    '90,000-99,999': '90,000-99,999',
    '100,000-124,999':'100,000-124,999',
    '125,000-149,999':'125,000-149,999',
    '150,000-199,999':'150,000-199,999',
    '200,000-249,999':'200,000-299,999',
    '250,000-299,999':'200,000-299,999', 
    '300,000-499,999':'>300,000',
    '$500,000-999,999':'>300,000',
    '>$1,000,000':'>300,000'}
```

```python
salary_encode = {   '$0-999': 0,
    '1,000-1,999': 0,
    '2,000-2,999': 0, 
    '3,000-3,999': 0,
    '4,000-4,999': 0,
    '5,000-7,499': 0, 
    '7,500-9,999': 0,
    '10,000-14,999': 1,
    '15,000-19,999': 1, 
    '20,000-24,999': 2,
    '25,000-29,999': 2,
    '30,000-39,999': 3, 
    '40,000-49,999': 4, 
    '50,000-59,999': 5,
    '60,000-69,999': 6,
    '70,000-79,999': 7, 
    '80,000-89,999': 8, 
    '90,000-99,999': 9,
    '100,000-124,999':10,
    '125,000-149,999':11,
    '150,000-199,999':12,
    '200,000-249,999':13,
    '250,000-299,999':13,
    '300,000-499,999':14,
    '$500,000-999,999':14,
    '>$1,000,000':14}
```

```python
#Label Encoding the target variable
Salaries.loc[1:,'Q29_Encoded'] = Salaries.loc[1:,'Q29'].map(salary_encode)
Salaries.loc[1:,'Q29_Encoded']=Salaries.loc[1:,'Q29_Encoded'].astype(int)
Salaries.Q29_Encoded.unique()
```

```python
Salaries[Salaries['Q29_Encoded'].isna()]
```

```python
#Combining the salary buckets
Salaries.loc[1:,'Q29_buckets'] = Salaries.loc[1:,'Q29'].map(salary_buckets)
Salaries.Q29_buckets.unique()
```
